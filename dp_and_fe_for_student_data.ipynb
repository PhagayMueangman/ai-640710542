```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student.csv")
```

```python
df
```

```python
# ตรวจสอบว่ามี missing value หรือไม่ (ได้ True/False)
df.isnull().values.any()
```

```python
# ดูว่าคอลัมน์ไหนมี missing และมีจำนวนเท่าไหร่
df.isnull().sum()
```

```python
# แทนค่า missing ใน Numeric
df['อายุ'] = df['อายุ'].fillna(df['อายุ'].median())
df['GPA'] = df['GPA'].replace({-1: np.nan})
df['GPA'] = df['GPA'].fillna(df['GPA'].mean())
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(df['ส่วนสูง_cm'].median())
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(df['ชั่วโมงเรียน_ต่อสัปดาห์'].median())
```

**อธิบายเหตุผลของบรรทัด**


1. df['อายุ'] = df['อายุ'].fillna(df['อายุ'].median())

เติมข้อมูลที่หายไป (missing value/NaN) ในคอลัมน์ "อายุ" ด้วยมัธยฐาน (median) ของคอลัมน์นั้น
เหตุผล: อายุเป็นข้อมูลที่อาจมี outlier หรือไม่กระจายปกติ เช่น ส่วนใหญ่ 18–22 ปี แต่มีบางคน 35 ปี หากเติม mean อาจทำให้ค่าสูง/ต่ำเกินจริง การใช้ median จะคำนวณจากค่ากลางของข้อมูลที่มีอยู่ จึงเหมาะสมกว่า


2. df['GPA'] = df['GPA'].replace({-1: np.nan})

แทนค่าที่มีค่าเป็น -1 ด้วย NaN (missing value)
เหตุผล: บางครั้งข้อมูลอาจถูกกรอกค่าผิด หรือใช้ -1 แทนค่าที่ไม่มีข้อมูลจริง (missing) การแปลง -1 เป็น NaN จะช่วยให้รู้ว่าค่าไหนที่หายไปจริง ๆ และจัดการต่อได้ง่ายขึ้น


3. df['GPA'] = df['GPA'].fillna(df['GPA'].mean())

เติมข้อมูล GPA ที่ว่าง (NaN) ด้วยค่าเฉลี่ย (mean)
เหตุผล: GPA มักกระจายเป็นปกติ ไม่ค่อยมี outlier มาก ค่าเฉลี่ยจึงสะท้อนความเป็นไปได้ของข้อมูลนักศึกษาโดยรวมได้ดี เป็นวิธีเติมข้อมูลที่ว่างที่นิยมใช้กับข้อมูลลักษณะนี้


4. df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(df['ส่วนสูง_cm'].median())

เติมส่วนสูงที่ว่าง (NaN) ด้วยค่า median (มัธยฐาน)
เหตุผล: ค่าส่วนสูงอาจมี outlier ได้ (เช่นบางคนกรอกผิดเป็น 30 หรือ 300 เซนติเมตร) การเติม median จะลดผลกระทบของ outlier และสะท้อนค่ากลางของข้อมูลส่วนสูงได้ดีกว่า mean


5. df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(df['ชั่วโมงเรียน_ต่อสัปดาห์'].median())

เติม "ชั่วโมงเรียนต่อสัปดาห์" ที่ว่าง (NaN) ด้วยมัธยฐาน (median)
เหตุผล: ชั่วโมงเรียนอาจมีค่าผิดปกติได้ เช่นบางคนติ๊ก 0 หรือ 50 ชั่วโมง หากใช้ mean แล้วอาจเพี้ยน การใช้ median จะสะท้อนค่าทั่วไปของผู้เรียนมากที่สุด


# **สรุปใจความ**
การเติม missing value (imputation) ด้วยมัธยฐาน (median) เหมาะกับข้อมูลที่มีโอกาสเกิด outlier เพื่อป้องกันการบิดเบือนของข้อมูลรวม
การเติมค่าด้วย mean (ค่าเฉลี่ย) นิยมใช้กับข้อมูลที่กระจายเป็นปกติและไม่มี outlier รุนแรง เช่น GPA
การแปลงค่าที่ไม่ปกติ (-1) ให้เป็น NaN ก่อน เป็นวิธีที่ช่วยให้จัดการค่าที่ผิดปกติหรือใช้แทน missing ได้ถูกต้อง

```python
# แทนค่า missing ใน Categorical
df['เพศ'] = df['เพศ'].fillna('ไม่ระบุ')
df['คณะ'] = df['คณะ'].fillna('ไม่ระบุ')
```

# **อธิบายทีละบรรทัด**

1. df['เพศ'] = df['เพศ'].fillna('ไม่ระบุ')

ทำหน้าที่แทนค่าที่ขาดหาย (missing value หรือ NaN) ในคอลัมน์ "เพศ" ด้วยคำว่า "ไม่ระบุ"
เหตุผล: การแทนค่าด้วย "ไม่ระบุ" ทำให้ข้อมูลสมบูรณ์ ไม่ต้องลบบรรทัดออก และแยกแยะได้ว่าแถวนี้ไม่มีการระบุเพศจริง ๆ (แทนที่จะเดาว่าชายหรือหญิง) ซึ่งช่วยลดอคติและไม่บิดเบือนการวิเคราะห์ข้อมูล


2. df['คณะ'] = df['คณะ'].fillna('ไม่ระบุ')

ทำหน้าที่แทนค่าที่ขาดหาย (NaN) ในคอลัมน์ "คณะ" ด้วย "ไม่ระบุ"
เหตุผล: เช่นเดียวกันกับคอลัมน์ "เพศ" การแทนค่าด้วย "ไม่ระบุ" ช่วยให้สามารถเก็บข้อมูลครบทุกแถว และหลีกเลี่ยงการกำหนดค่าหมวดหมู่อื่นแทน (ซึ่งอาจสร้างอคติ) อย่างถูกต้อง โดยข้อมูลนี้จะมีประโยชน์หากต้องแยกวิเคราะห์ภายหลัง
ทำไมจึงไม่ควรเติมด้วยค่าที่พบบ่อยที่สุด (mode) ในกรณีนี้?
ถ้าเติมด้วยค่า mode (เช่น "ชาย"/"หญิง" หรือ "วิทยาศาสตร์") จะทำให้ข้อมูลเกิดความเอนเอียง เพราะเราไปสมมุติค่านั้นแทนซึ่งอาจไม่ถูกต้องจริง ๆ
เติม "ไม่ระบุ" ทำให้ข้อมูลเปิดเผยสถานะที่แท้จริงของการขาดหาย และสามารถเลือกวิเคราะห์หรือกรองเฉพาะข้อมูลที่ครบถ้วนได้ (เช่น วิเคราะห์แต่แถวที่ระบุเพศหรือคณะ)
# **สรุป**
การเติม "ไม่ระบุ" ในคอลัมน์ข้อมูลประเภท Categorical ช่วยรักษาความสมบูรณ์ของข้อมูล
ลดความลำเอียง (bias)
ไม่ทำให้ขนาดข้อมูลลดลงโดยไม่จำเป็น

```python
# แทนค่าวันที่
df['วันที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].fillna(df['วันที่ลงทะเบียน'].mode()[0])
```

# **อธิบาย:**

  โค้ดนี้ใช้สำหรับเติมค่าที่หายไป (missing value/NaN) ในคอลัมน์ “วันที่ลงทะเบียน” ด้วย "mode" หรือค่าที่เกิดขึ้นบ่อยที่สุดในคอลัมน์นี้ (mode()[0] หมายถึงค่าที่พบบ่อยที่สุดลำดับแรก)


# **เหตุผลที่เหมาะสม:**
"วันที่ลงทะเบียน" มักหมายถึงวันที่นิสิตหรือนักศึกษาสมัครเรียนหรือกรอกข้อมูล ซึ่งโดยทั่วไปการเลือก “mode” เป็นวิธีที่ดีสำหรับข้อมูลวันที่กรณีที่วันที่เหล่านั้นจะเหมือนหรือคล้ายกัน เช่น ส่วนมากคนจะลงทะเบียนพร้อมกัน ถ้าใช้ค่าเฉลี่ย (mean) กับวันที่จะไม่ได้ใช้งานง่ายเหมือนค่าประเภทตัวเลข และ median ของวันที่ก็อาจเข้าใจยากหรือตีความลำบาก Mode คือวันที่ที่คนส่วนใหญ่ลงทะเบียนจริง จึงเหมาะที่จะเติมสำหรับค่า missing
วิธีนี้จะลดอคติ (bias) ที่จะเกิดจากการเดา หรือเติมวันที่ใหม่ที่ไม่มีในชุดข้อมูล

# **กรณีข้อมูลวันที่: เติมช่องว่างด้วยวันที่ที่พบบ่อยที่สุด (mode) เพราะวันที่เฉลี่ยหรือค่าว่างจะไม่มีความหมาย**

ยิ่งถ้าสัดส่วน missing มีน้อย การลบทิ้งจะกระทบข้อมูลส่วนน้อยเท่านั้น
